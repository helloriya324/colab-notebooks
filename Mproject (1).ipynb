```python
#MACHINE LEARNING - SUPERVISED LEARNING - REGRESSION - LINEAR REGRESSION
#Univariate/Single- 1 column as input and 1 column as output
#Multivariate/Multiple- multiple column as input and 1 column as output
#Dataset
#Year in numbers, Salary in rupees
```

```python
#1. Take the data and create dataframe
import pandas as pd
df= pd.read_csv('/content/Salary_Data.csv')
df
```

```python
#2. We are not performing step no 2, for this dataset
```

```python
#3. DATA VISUALISATION- CREATION OF GRAPHS
import matplotlib.pyplot as plt
plt.scatter(df['YearsExperience'],df['Salary']) #plt.scatter(x axis, y axis)
plt.title('YearsExperience vs Salary')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
```

```python
#INPUT- YearsExperience
#OUTPUT- Salary
#4. DIVIDE THE DATA INTO INPUT AND OUTPUT
#INPUT(x) is always 2d array, output(y) is always 1d array
x= df.iloc[0:6,0:1].values #df.iloc[row slicing, column slicing]  #.values converts the dataframe into an array
x
```

```python
y= df.iloc[0:6,1].values  #if we want all the column we will put only :
y
```

```python
#5.TRAIN AND TEST VARIABLES
#Here we are not performing step 5 due to limited data
```

```python
#6. NORMALISATION/SCALING
#IT IS ONLY DONE FOR MULTIVARIATE DATASETS
#Here for UNIVARIATE data, step no 6 is not required
```

```python
#7. Run a classifier,REGRESSOR or clusterer(APPLYING a suitable ALGORITHM)
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#8. FIT the MODEL(mapping/plotting the inputs with the outputs in the library)
#LinearRegression.fit(x,y)
model.fit(x,y)
```

```python
#9. PREDICT THE OUTPUT
y_pred= model.predict(x) #we predict the output
y_pred #predicted output values
```

```python
y  #actual outpuut values
```

```python
#CONCLUSION: We have to compare the corresponding values of y and y_pred.
#when we compare we came to know that there is a huge diff . This doesn't mean that our model has predicted wrong.
#It only means that our model is NOT LINEAR/LESS LINEAR
#LINEARITY of a model depends upon the NATURE of the data as well as the SIZE of the data
```

```python
#INDIVIDUAL PREDICTION
#I want to know the salary for 1.2years
model.predict([[1.2]]) 
```

```python
#CROSS VERIFICATION
#y= mx+c #eqn of a straight line
#m= slope
#c= const
#y= dependent variable
#x= independent variable

#From excel m=9020.6356, c=28216.868
```

```python
m= model.coef_   #slope(m)
m
```

```python
c= model.intercept_
c
```

```python
#y= mx+c 
m*1.2 +c 
```

```python
#FINAL VISUALISATION- BEST FIT LINE

plt.scatter(x,y)  #actual values
plt.plot(x,y_pred,c = 'red') #predicted values
plt.title('BEST FIT LINE')
plt.xlabel('YearsExperience')
plt.ylabel('Salary') 
```
